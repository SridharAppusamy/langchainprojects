```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_huggingface import HuggingFaceEndpoint
```

```python
repo_id="mistralai/Mistral-7B-Instruct-v0.3"
llm=HuggingFaceEndpoint(repo_id=repo_id,temperature=0.7,huggingfacehub_api_token=os.getenv("HF_TOKEN"),max_new_tokens=150)
llm
```

```python
llm.invoke("what is Machine Learning")
```

```python
from langchain import PromptTemplate,LLMChain
template="""
Question:{question}
answer:lets thing step by step
"""
prompt=PromptTemplate(
    input_variables=['question'],
    template=template
)
print(prompt)
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
llm_chain.invoke({"question":"what is machine learning"})
```

```python
llm_chain.invoke("what is machine learning")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': False}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
text = "This is text"
query_result = hf.embed_query(text)
query_result
```
